```python
#파일 불러오기
file = open('Atmosphere.txt', 'r')
```

```python
data = file.read()
```

```python
#형태소 단위로 분해
from konlpy.tag import Okt
from collections import Counter
okt = Okt()
morphs = okt.morphs(data)
```

#길이가 1보다 큰 단어들만 뽑아내기
words = []
for n in morphs:
    if len(n) > 1:
        words.append(n)

```python
#빈도수가 가장 높은 단어 300개 추출
count = Counter(morphs)
morph_list = count.most_common(300)
```

```python
file2 = open('title_2019_times.txt', 'w')
file2.writelines(str(morph_list))
```

```python
from pandas import DataFrame
news_title = DataFrame(morph_list) #키워드 빈도수 정리
print(news_title)
file3 = open('title_2019_noun.txt', 'w')
file3.writelines(str())
```

```python
file.close()
file2.close()
file3.close()
```

```python
file = open('title_2019.txt', 'r')
message = file.read()
```

```python
import numpy as np
from PIL import Image
from wordcloud import WordCloud
#배경으로 2019 사용
icon = Image.open('2019.jpeg').convert('RGBA')
mask = Image.new('RGB', icon.size, (255,255,255))
mask.paste(icon,icon)
mask = np.array(mask)
```

```python
# 색 지정(인터넷 검색 더 해봐야 할듯),,
from matplotlib.colors import LinearSegmentedColormap
colors = ["#c6a554", "#5b1a20", "#ceaea3", "#1f3238", "#48372d"]
#https://www.color-hex.com/color-palettes/ (색 찾아서 사용)
cmap = LinearSegmentedColormap.from_list('mycmap', colors)
```

```python
wc = WordCloud (font_path = '/System/Library/Fonts/Supplemental/AppleGothic.ttf', background_color = 'white',
                max_words = 400, mask = mask, max_font_size = 500, random_state = 42, colormap = cmap).generate_from_text(message)
```

```python
wc.to_file('title_2019.png')
```
