```python
file = open('title_2020.txt', 'r')
message = file.read()
```

```python
import numpy as np
from PIL import Image
from wordcloud import WordCloud
#배경으로 2019 사용
icon = Image.open('2020.png').convert('RGBA')
mask = Image.new('RGB', icon.size, (255,255,255))
mask.paste(icon,icon)
mask = np.array(mask)
```

```python
# 색 지정(인터넷 검색 더 해봐야 할듯),,
from matplotlib.colors import LinearSegmentedColormap
colors = ["#c6a554", "#5b1a20", "#ceaea3", "#1f3238", "#48372d"]
#https://www.color-hex.com/color-palettes/ (색 찾아서 사용)
cmap = LinearSegmentedColormap.from_list('mycmap', colors)
```

```python
wc = WordCloud (font_path = '/System/Library/Fonts/Supplemental/AppleGothic.ttf', background_color = 'white',
                max_words = 400, mask = mask, max_font_size = 500, random_state = 42, colormap = cmap).generate_from_text(message)
```

```python
wc.to_file('title_2020.png')
```
