```python
import pandas as pd
df = pd.read_csv('pollution_dataset(2017-2020).csv')
df
#print(df['오존농도(ppm)'].min(),df['오존농도(ppm)'].max())
```

```python
import numpy as np
df['오존농도(ppm)'].dtype
```

```python
#결측치 일단 drop
df.dropna(subset = ['오존농도(ppm)'], axis = 0, inplace = True)
count, bin_dividers = np.histogram(df['오존농도(ppm)'], bins = 4)
print(bin_dividers)
```

```python
import pandas as pd
bin_names = [1, 2, 3, 4]
df['오존농도(ppm)구간'] = pd.cut(x = df['오존농도(ppm)'],
                           bins = bin_dividers,
                           labels = bin_names,
                           include_lowest = True)
df[['오존농도(ppm)', '오존농도(ppm)구간']].tail(15)
print(df['오존농도(ppm)구간'].describe())
```
